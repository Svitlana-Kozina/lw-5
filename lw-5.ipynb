#### Побудувати **кожну лінію на окремому графіку**, розмістивши їх поруч:
- всі прямі повинні бути різного кольору, але не використовувати
системну послідовність кольорів (синій, помаранчевий, зелений);
- всі прямі повинні мати різний тип ліній (пунктирна, точка тире тощо);
- всі графіки повинні мати сумісну вісь ординат;
- кожен графік повинен мати підпис.

```python
import numpy as np
import matplotlib.pyplot as plt

# Дані: коефіцієнти для трьох ліній
# 10.2x1 - 3.2x2 = 149  →  x2 = (10.2x1 - 149) / 3.2
# -5.8x1 + 16x2 = 83    →  x2 = (5.8x1 + 83) / 16
# 10.3x1 + 7.3x2 = 234  →  x2 = (234 - 10.3x1) / 7.3

# np.linspace(start, stop, num) — генерує рівномірно розподілені точки.
# х1 це NumPy-масив, тип float
x1 = np.linspace(-10, 20, 100)

# Виражаємо у х2 через х1
y1 = (10.2 * x1 - 149) / 3.2
y2 = (5.8 * x1 + 83) / 16
y3 = (234 - 10.3 * x1) / 7.3

# Налаштування фігури: 3 графіки поруч
# plt.subplots() створює одразу всю фігуру (тобто вікно з графіками) і підграфіки
# plt.subplots() повертає два об’єкти: fig — це фігура, axes — це набір осей.
# 1, 3 → 1 рядок, 3 колонки (тобто 3 графіки поруч); 
# figsize=(15, 4) → розмір зображення (15×4 дюйми); 
# sharey=True → спільна вісь Y (ординат) для всіх трьох графіків (однаковий масштаб по вертикалі);
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

# Власні (не системні) кольори
colors = ['#41d363', '#8700f2', '#ff7f50']

# Різні стилі ліній
linestyles = ['--', ':', '-.']

# Підписи для кожного графіка
# r'...' сирий рядок, передати як є, Python НЕ обробляє спецсимволи
titles = [
    r'$10.2x_1 - 3.2x_2 = 149$',
    r'$-5.8x_1 + 16x_2 = 83$',
    r'$10.3x_1 + 7.3x_2 = 234$'
]

# Побудова трьох окремих графіків
axes[0].plot(x1, y1, color=colors[0], linestyle=linestyles[0])
axes[1].plot(x1, y2, color=colors[1], linestyle=linestyles[1])
axes[2].plot(x1, y3, color=colors[2], linestyle=linestyles[2])

# Додавання підписів і осей
for i in range(3):
    axes[i].set_title(titles[i], fontsize=12)
    axes[i].set_xlabel('$x_1$')
    axes[i].grid(True)

# Спільна вісь ординат (y), х2
axes[0].set_ylabel('$x_2$')

plt.tight_layout()

plt.savefig('figure1.jpg', format='jpg', dpi=300, bbox_inches='tight')
plt.savefig('figure1.png', format='png', dpi=300, bbox_inches='tight')
plt.savefig('figure1.svg', format='svg', bbox_inches='tight')

plt.show()
```

#### Розмістити **всі лінії на одному** рисунку:
- всі прямі повинні бути різного кольору, але не використовувати
системну послідовність кольорів (синій, помаранчевий, зелений);
- всі прямі повинні мати різний тип ліній (пунктирна, точка тире тощо);
- підібрати масштаб таким чином, щоб всі три точки перетину прямих
були в області видимості;
- додати легенду на графік;
- додати сітку, в якій задати колір та тип ліній;
- змінити розмір рисунку (наприклад, 8х16 дюймів) та розподільчу
здатність (наприклад, 100 dpi);
- зробити підписи рівнянь прямих вздовж лінії з відповідним нахилом;
- додати підписи осей та назву графіку;
- заповнити кольором область, що утворена перетином всіх прямих.

За допомогою підмодуля numpy.linalg **знайти точки перетину всіх пар
прямих, відмітити їх** та зробити відповідні вказівки на графіку.

```python
import numpy as np
import matplotlib.pyplot as plt

# 10.2x1 - 3.2x2 = 149  →  x2 = (10.2x1 - 149) / 3.2
# -5.8x1 + 16x2 = 83    →  x2 = (5.8x1 + 83) / 16
# 10.3x1 + 7.3x2 = 234  →  x2 = (234 - 10.3x1) / 7.3

# Розширили діапазон, щоб бачити всі перетини
# Створює 400 рівновіддалених точок від -10 до 30
x1 = np.linspace(-10, 30, 400) 

# y1, y2 і y3 — це масиви NumPy (numpy.ndarray)
# масив із 400 результатів обчислення y для кожного x1
y1 = (10.2 * x1 - 149) / 3.2
y2 = (5.8 * x1 + 83) / 16
y3 = (234 - 10.3 * x1) / 7.3

# Створення фігури (вікна для графіка)
# Налаштування фігури figsize=(ширина, висота)
plt.figure(figsize=(8, 8), dpi=100)  # 8x8 дюймів, 100 dpi

# Кольори ліній
colors = ['#a855ff', '#53BDA5', '#ff4162']

# Різні стилі ліній
linestyles = [':', '-.', '--']

# Побудова трьох ліній на одному графіку
plt.plot(x1, y1, color=colors[0], linestyle=linestyles[0], linewidth=2, label=r'$10.2x_1 - 3.2x_2 = 149$')
plt.plot(x1, y2, color=colors[1], linestyle=linestyles[1], linewidth=2, label=r'$-5.8x_1 + 16x_2 = 83$')
plt.plot(x1, y3, color=colors[2], linestyle=linestyles[2], linewidth=2, label=r'$10.3x_1 + 7.3x_2 = 234$')

# Обчислення точок перетину (для масштабу й області)
# NumPy-масив (матриця) A: коефіцієнти при x1 і x2 у кожному рівнянні
A = np.array([[10.2, -3.2], [-5.8, 16], [10.3, 7.3]])
#b[i] - вільний член (права частина) рівняння
b = np.array([149, 83, 234])

# Щоб знайти точки, де лінії графіків перетинаються, 
# треба розв’язати по два рівняння з них одночасно.
# Три рівняння → три пари: 1-ше і 2-ге, 1-ше і 3-тє, 2-ге і 3-тє.
# Отже, буде 3 точки перетину.

# Функцію-комбінатор автоматично створює всі пари елементів рівнянь без повторів
# для знаходження точок перетину.
from itertools import combinations

# Порожній список, у який ми будемо складати 
# знайдені точки перетину, координати х,у, у нас це х1,х2
# кожен елемент — це [x₁, x₂]
points = []

# Утворює всі пари індексів трьох рівнянь: (0, 1), (0, 2), (1, 2)
for (i, j) in combinations(range(3), 2):
    Aij = A[[i, j], :] # беремо з матриці A лише два рядки
    bij = b[[i, j]]    # беремо 2 відповідних значення b (вільний член без х, права ч.рівняння)
    
    # Розв’язуємо систему з двох рівнянь і отримуємо координати точки перетину, sol = [x1, x2]
    sol = np.linalg.solve(Aij, bij)

    # Додаємо знайдену точку у список points
    points.append(sol)

# Перетворюємо список на масив
points = np.array(points)

# Відмітити точки перетину на графіку
plt.scatter(points[:, 0], points[:, 1], color='black', s=60, zorder=5, label='Точки перетину')

# Додати підписи до кожної точки
# Індивідуальні зсуви для кожної точки
offsets = [(0.3, 0.7), (-1.5, 2.5), (2.5, -3)]

for i, ((x, y), (dx, dy)) in enumerate(zip(points, offsets)):
    plt.text(x + dx, y + dy,
             f'P{i+1}({x:.1f}, {y:.1f})',
             fontsize=10, color='black',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# З’єднує всі ці точки х1 х2 і зафарбовує всередині
plt.fill(points[:, 0], points[:, 1], color='black', alpha=0.2, label='Область перетину')

# Підписи рівнянь уздовж ліній
plt.text(10, (10.2*15 - 149)/3.2 - 12, r'$10.2x_1 - 3.2x_2 = 149$', color=colors[0], rotation=63)
plt.text(5, (5.8*5 + 83)/16 + 1.5, r'$-5.8x_1 + 16x_2 = 83$', color=colors[1], rotation=12)
plt.text(5, (234 - 10.3*10)/7.3 + 2, r'$10.3x_1 + 7.3x_2 = 234$', color=colors[2], rotation=-40)

# Сітка, осі, легенда, підписи
plt.grid(True, color='gray', linestyle=':', linewidth=0.8)
plt.xlabel('$x_1$', fontsize=12)
plt.ylabel('$x_2$', fontsize=12)
plt.title('Перетин трьох прямих', fontsize=14)

#loc='best' - розмістити легенду там, де вона найменше перекриває графік
plt.legend(loc='best')

# Масштаб, початок та кінець шкали по х та по у
plt.xlim(-5, 25)
plt.ylim(-20, 30)

# Автоматично підганяє відступи між усіма елементами
plt.tight_layout()

plt.savefig('figure2.jpg', format='jpg', dpi=300, bbox_inches='tight')
plt.savefig('figure2.png', format='png', dpi=300, bbox_inches='tight')
plt.savefig('figure2.svg', format='svg', bbox_inches='tight')

# Відображає графік на екрані, графік точно з'явиться на екрані у різному ПЗ
plt.show()
```

**Формат .jpg** має втрати якості через стиснення — на зображенні з’являються артефакти та легке розмиття ліній і тексту. Растрові зображення, які зберігається у вигляді пікселів.

**Формат .png** зберігає зображення більш чіткими. Растровий формат, підтримує прозорість фону.

**Формат .svg** можна масштабувати без втрати якості, а розмір файлу залишається невеликим. Векторний формат, зберігає графіку як математичні лінії та фігури.

Таким чином, .jpg підходить для фотографій, .png — для звітів і презентацій, а .svg — для професійної публікації або веб-графіки.
